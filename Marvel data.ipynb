```python
import pandas as pd
import matplotlib.pyplot as plt

mcu_movies = pd.read_csv('C:\\Users\\saman\\OneDrive\\Documents\\Coding Temple Work\\mcu.csv')

print(mcu_movies)
```

```python
mcu_movies['Release Date'] = pd.to_datetime(mcu_movies['Release Date'])

plt.figure(figsize=(10, 6))
plt.plot(mcu_movies['Release Date'], mcu_movies['Inflation Adjusted Domestic'], marker='o')
plt.title('Inflation-Adjusted Domestic Box Office Performance of Marvel Movies')
plt.xlabel('Release Date')
plt.ylabel('Inflation-Adjusted Domestic Box Office (USD)')
plt.xticks(rotation=90)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
mcu_movies['Release Date'] = pd.to_datetime(mcu_movies['Release Date'])

plt.figure(figsize=(10, 6))
plt.plot(mcu_movies['Release Date'], mcu_movies['Opening Weekend'], marker='o')
plt.title('Opening Weekend Performance of Marvel Movies')
plt.xlabel('Release Date')
plt.ylabel('Opening Weekend Box Office (USD)')
plt.xticks(rotation=90)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#This data shows that there is a fairly strong correlation between the imbd score and how well the movie does in the box office, the rotten tomato Audience score seems to have little to no correlation
correlation_imdb = mcu_movies['IMDb Score'].corr(mcu_movies['Inflation Adjusted Domestic'])

correlation_rotten_tomato = mcu_movies['Rotten Tomato Audience Score'].corr(mcu_movies['Inflation Adjusted Domestic'])
correlation_tomatometer = mcu_movies['Tomatometer'].corr(mcu_movies['Inflation Adjusted Domestic'])
print("Correlation between IMDb Score and Inflation-Adjusted Domestic Box Office:", correlation_imdb)
print("Correlation between Tomatometer score and Inflation-Adjusted Domestic Box Office:", correlation_tomatometer)
print("Correlation between Rotten Tomato Audience Score and Inflation-Adjusted Domestic Box Office:", correlation_rotten_tomato)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(mcu_movies['Tomatometer'], mcu_movies['Inflation Adjusted Domestic'], color='skyblue', alpha=0.7)
plt.title('Correlation between Rotten Tomatoes Tomatometer Score and Inflation-Adjusted Domestic Box Office')
plt.xlabel('Rotten Tomatoes Tomatometer Score')
plt.ylabel('Inflation-Adjusted Domestic Box Office (USD)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(mcu_movies['IMDb Score'], mcu_movies['Inflation Adjusted Domestic'], color='skyblue', alpha=0.7)
plt.title('Correlation between IMDb Score and Inflation-Adjusted Domestic Box Office')
plt.xlabel('IMDb Score')
plt.ylabel('Inflation-Adjusted Domestic Box Office (USD)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
correlation_budget = mcu_movies['Budget'].corr(mcu_movies['Inflation Adjusted Domestic'])

print("Correlation between Budget and Inflation-Adjusted Domestic Box Office:", correlation_budget)
```

```python
phase_means = mcu_movies.groupby('Phase')['Inflation Adjusted Domestic'].mean().sort_values()

plt.figure(figsize=(10, 6))
phase_means.plot(kind='bar', color='skyblue')
plt.title('Mean Inflation-Adjusted Domestic Box Office Performance by Phase')
plt.xlabel('Phase')
plt.ylabel('Mean Inflation-Adjusted Domestic Box Office (USD)')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
mcu_movies['Release Date'] = pd.to_datetime(mcu_movies['Release Date'])

mcu_movies_sorted = mcu_movies.sort_values(by='Release Date')

mcu_movies_sorted['Release Frequency (Days)'] = mcu_movies_sorted['Release Date'].diff().dt.days

print(mcu_movies_sorted[['Movie', 'Release Date', 'Release Frequency (Days)']])
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(mcu_movies_sorted['Release Frequency (Days)'], mcu_movies_sorted['Inflation Adjusted Domestic'], color='skyblue', alpha=0.7)
plt.title('Release Frequency vs. Inflation-Adjusted Domestic Box Office')
plt.xlabel('Release Frequency (Days)')
plt.ylabel('Inflation-Adjusted Domestic Box Office (USD)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
